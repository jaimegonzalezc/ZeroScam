```python
from datasets import load_dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    TrainingArguments,
    BitsAndBytesConfig
)
from peft import LoraConfig, prepare_model_for_kbit_training, get_peft_model
from trl import SFTTrainer
import torch
from huggingface_hub import notebook_login
```

Configuro el modelo y tokenizador. Utilizo el quantificador 4bit para ahorrar memoria.

```python
# Configurar modelo
model_name = "deepseek-ai/DeepSeek-R1-Distill-Llama-8B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)

# Configurar LoRA
peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = prepare_model_for_kbit_training(model)
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()
```

```python
# Cargar datasets de ciberseguridad
def download_and_standardize(dataset_name, rename_columns=None):
    dataset = load_dataset(dataset_name)
    data = dataset["train"] if "train" in dataset else dataset[list(dataset.keys())[0]]
    data_list = [dict(row) for row in data]

    if rename_columns:
        for row in data_list:
            for old_key, new_key in rename_columns.items():
                # Renombrar columnas si existen en el dataset
                if old_key in row:
                    row[new_key] = row.pop(old_key)

            # Asegurar que 'context' siempre esté presente
            if "context" not in row:
                row["context"] = "No context available"

    return data_list

datasets_info = [
    {"name": "ahmed000000000/cybersec", "rename_columns": {"INSTRUCTION": "question", "RESPONSE": "answer"}},
    {"name": "dzakwan/cybersec", "rename_columns": {"instruction": "context", "input": "question", "output": "answer"}},
    {"name": "asimsultan/cyber2k", "rename_columns": None},  # No se renombrarán columnas
    {"name": "Vanessasml/cybersecurity_32k_instruction_input_output", "rename_columns": {"instruction": "context", "input": "question", "output": "answer"}}
]

final_dataset = []
for dataset in datasets_info:
    final_dataset.extend(download_and_standardize(dataset["name"], dataset["rename_columns"]))

# Verificar estructura final
print(final_dataset[0])  # Muestra un ejemplo para confirmar que 'context' está presente
```

```python
from datasets import Dataset

hf_dataset = Dataset.from_list(final_dataset)
print(hf_dataset)  # Ahora sí puedes ver filas y columnas correctamente
```

```python
hf_dataset_train_test = hf_dataset.train_test_split(test_size=0.2, seed=420, shuffle=True)
hf_dataset_val_test = hf_dataset_train_test["test"].train_test_split(test_size=0.5, seed=420, shuffle=True)


split_hf_dataset = DatasetDict({
    "train": hf_dataset_train_test["train"],
    "val": hf_dataset_val_test["train"],
    "test": hf_dataset_val_test["test"]
})
print(split_hf_dataset)
```

```python
# Reducción de dataset
def sample_dataset_with_proportions(dataset_dict, total_sample_size):
    total_examples = sum(len(dataset_dict[split]) for split in dataset_dict.keys())
    sampled_dataset = DatasetDict()
    for split in dataset_dict.keys():
        split_proportion = len(dataset_dict[split]) / total_examples
        split_sample_size = min(int(total_sample_size * split_proportion), len(dataset_dict[split]))
        sampled_dataset[split] = dataset_dict[split].shuffle(seed=420).select(range(split_sample_size))
    return sampled_dataset

total_sample_size = 1000
reduced_dataset = sample_dataset_with_proportions(split_hf_dataset, total_sample_size)
```

```python
# Definir prompt de entrenamiento
def format_prompt(example):
    return f"""Eres un experto en ciberseguridad para usuarios no técnicos. Responde en español de forma clara y sencilla.

### Instrucción:
{example['question']}

### Contexto:
{example['context']}

### Respuesta:
{example['answer']}"""
```

```python
# Configurar entrenamiento
training_args = TrainingArguments(
    output_dir="./deepseek-ciberseguridad",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    gradient_checkpointing=True,
    learning_rate=2e-5,
    lr_scheduler_type="cosine",
    optim="adamw_torch_fused",
    bf16=True,
    logging_steps=20,
    save_strategy="steps",
    report_to="tensorboard"
)

trainer = SFTTrainer(
    model=model,
    processing_class=tokenizer,
    train_dataset=reduced_dataset["train"],
    eval_dataset=reduced_dataset["val"],
    formatting_func=format_prompt,
    args=training_args
)
```

```python
# Entrenar y guardar modelo
trainer.train()
trainer.model.save_pretrained("./deepseek-ciberseguridad-lora")
tokenizer.save_pretrained("./deepseek-ciberseguridad-lora")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

model_name = "CasiAC/deepseek-ciberseguridad-lora"

# Configuración para cargar en 8bit o 4bit
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,  # Cambia a load_in_4bit=True para menor consumo de VRAM
)

# Cargar el modelo con 8bit
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)

tokenizer = AutoTokenizer.from_pretrained(model_name)

print("Modelo cargado en 8bit con éxito 🚀")

```

```python
import torch

# Asegurar que el modelo y el tokenizador están cargados
assert "model" in globals(), "El modelo no está cargado. Carga el modelo antes de ejecutar este código."
assert "tokenizer" in globals(), "El tokenizador no está cargado. Carga el tokenizador antes de ejecutar este código."

# Definir dispositivo
device = "cuda" if torch.cuda.is_available() else "cpu"

# Configurar el token de padding (para evitar la advertencia)
tokenizer.pad_token = tokenizer.eos_token

# Definir el prompt de evaluación
prompt = """Eres un asistente de ciberseguridad. Explica de forma simple y en español:
¿Qué es un ataque de phishing y cómo puedo evitarlo?"""

# Tokenizar la entrada y moverla a la GPU
inputs = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True).to(device)

# Generar respuesta con mejor formato
outputs = model.generate(
    inputs.input_ids,
    attention_mask=inputs.attention_mask,  # Se añade la máscara de atención
    max_new_tokens=300,
    temperature=0.7,
    do_sample=True,
    top_k=50,
    top_p=0.9,
    repetition_penalty=1.1,
    pad_token_id=tokenizer.eos_token_id,  # Evita advertencias
)

# Decodificar y mostrar la respuesta formateada
response = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Ajustar el formato final
response = response.replace(prompt, "").strip()  # Eliminar el prompt de la respuesta

print(f"\n<start>\n{response}\n</start>")
```
